```python
df = pd.read_csv(r"Y:\code\regression-pipeline\data\raw\HouseTS.csv")
df.to_parquet(r"Y:\code\regression-pipeline\data\raw\HouseTS.parquet")
```

```python
df["date"] = pd.to_datetime(df["date"])
print(df["date"].dt.year.value_counts())
```

```python
# sorting by date
df = df.sort_values("date")
```

```python
cutoff_date_val = "2020-01-01" # validation data start
cutoff_date_holdout = "2022-01-01" # holdout data start (not touched at all)
```

```python
# splitting the dataframe
train_df = df[df["date"] < cutoff_date_val]

# validation - eval : 2020 - 2021
val_df = df[(df["date"] >= cutoff_date_val) & (df["date"] < cutoff_date_holdout)]

# holdout : 2022 - 2023
holdout_df = df[df["date"] >= cutoff_date_holdout]

print("Train shape", train_df.shape)
print("Val shape", val_df.shape)
print("Holdout shape", holdout_df.shape)
```

```python
# saving splits
train_df.to_csv(r"Y:\code\regression-pipeline\data\raw\train.csv", index=False)
val_df.to_csv(r"Y:\code\regression-pipeline\data\raw\val.csv", index=False)
holdout_df.to_csv(r"Y:\code\regression-pipeline\data\raw\holdout.csv", index=False)
```
