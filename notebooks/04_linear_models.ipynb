## Linear Regression Model

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
# loading dfs'
train_df = pd.read_csv(r"Y:\code\regression-pipeline\data\processed\fe_train.csv")
val_df = pd.read_csv(r"Y:\code\regression-pipeline\data\processed\fe_val.csv")
```

```python
target = "price"
X_train = train_df.drop(columns=[target])
Y_train = train_df[target]

X_val = val_df.drop(columns=[target])
Y_val = val_df[target]
```

```python
# Standardization (fit on train, transform eval)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

```python
# Linear Regression
lr = LinearRegression()
lr.fit(X_train_scaled, Y_train)
Y_pred_lr = lr.predict(X_val_scaled)

print("Linear Regression:")
print(f"MAE {mean_absolute_error(Y_val, Y_pred_lr)}")
print(f"RMSE {np.sqrt(mean_squared_error(Y_val, Y_pred_lr))}")
print(f"R² {r2_score(Y_val, Y_pred_lr)}")
```

```python
# Ridge Regression
ridge = Ridge(alpha=1.0)
ridge.fit(X_train_scaled, Y_train)
Y_pred_ridge = ridge.predict(X_val_scaled)


print("Ridge Regression:")
print(f"MAE {mean_absolute_error(Y_val, Y_pred_ridge)}")
print(f"RMSE {np.sqrt(mean_squared_error(Y_val, Y_pred_ridge))}")
print(f"R² {r2_score(Y_val, Y_pred_ridge)}")
```

```python
# Lasso Regerssion
lasso = Lasso(alpha=0.1)
lasso.fit(X_train_scaled, Y_train)
y_pred_lasso = lasso.predict(X_val_scaled)

print("\nLasso Regression:")
print(" MAE:", mean_absolute_error(Y_val, y_pred_lasso))
print(" RMSE:", np.sqrt(mean_squared_error(Y_val, y_pred_lasso)))
print(" R²:", r2_score(Y_val, y_pred_lasso))
```

```python
# ElasticNet
elastic = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic.fit(X_train_scaled, Y_train)
y_pred_elastic = elastic.predict(X_val_scaled)

print("\nElasticNet Regression:")
print(" MAE:", mean_absolute_error(Y_val, y_pred_elastic))
print(" RMSE:", np.sqrt(mean_squared_error(Y_val, y_pred_elastic)))
print(" R²:", r2_score(Y_val, y_pred_elastic))
```
